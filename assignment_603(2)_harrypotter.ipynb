## 1. Write Python code and use MapReduct to count occurrences of each word in the first text file (file.txt). How many times each word is repeated?


<h1> MY DOB IS 24/SEP/1998 <h1>
    
 <h4> the month is september(09), it is divided by 2 and rounded to 5. the 5th book is Harry Potter and the order of phonex <h4>
 <h4> the date is 24 so I have extracted the 10 pages starting from the 24th page (file1) <h4> 
 <h4>. the year is 98, so i have extracted the 10 pages starting from 98th (page file2) <h4>    

```python
import re                     # importing the required packages
import string
from functools import reduce
from collections import Counter

```

```python
def map_words(text):
    words = re.findall(r'\w+', text.lower()) #  regular expression '\w+' takes whole words consisting of word characters, including letters, digits, and underscores.
    return words                                       # returning the list of words
```

```python
with open('file1.txt', 'rb') as file:                 # Opening the 'file1.txt' in binary read mode 
    lines = file.readlines()                          # Reading all lines from the file
    lines = [line.decode('utf-8') for line in lines]
    text = ''.join(lines)                             # Joining the list of lines to form a single  string
    words = map_words(text)                           # calling the 'map_words' function whhich split the text into individual words

```

```python
word_counts_list = list(map(Counter, [words]))       # Using the map function and 'Counter' class to count the occurrences of each word in the 'words' list.
total_word_counts = reduce(lambda x, y: x + y, word_counts_list) # Using 'reduce' function to combine the word counts from different portions into a single total word count.

for word, count in total_word_counts.items():   # Iterateing through the total word counts and printing each word and its count.
    print(f'{word}: {count}')
```

### From the second text file (file2.txt), write Python code and use MapReduct to count how many times non-English words (names, places, spells etc.) were used. List those words and how many times each was repeated.

```python
import re                                  # importing the required packages
from functools import reduce
from spellchecker import SpellChecker

```

```python
def map_words(text):
    return re.findall(r'\w+', text.lower())  # regular expression '\w+' take one or more word characters, which include letters and underscores.
```

```python
def is_non_english_word(word, spellchecker):
    return not word.isdigit() and word not in spellchecker   # Checking if a word is non-English
                                                             


```

```python


spellchecker = SpellChecker()                               # Initializing a SpellChecker instance to check for English words.
word_counts = {}                                            # Initializing a dictionary to store word counts.

with open('file2.txt', 'r', encoding='utf-8') as file:      # Opening 'file2.txt' in UTF-8 encoding for reading.
    text = file.read()                                      # Reading the entire content of the file into the 'text' variable.
    words = map_words(text)                                 # calling the 'map_words' function to split the 'text' into individual words.

    for word in words:                                      # Iterating through each word in 'words'.
        if is_non_english_word(word, spellchecker):         # Checking if the word is non-English or english
            word_counts[word] = word_counts.get(word, 0) + 1 # If the word is not in 'word_counts', initialize its count to 1.

for word, count in word_counts.items():                     # Iterating through the 'word_counts' dictionary and printing each word and its count.
    print(f'{word}: {count}')
```
